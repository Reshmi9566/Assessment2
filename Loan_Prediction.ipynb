```python
cat_cols = [col for col in df1.columns if df1[col].dtype == 'object']
num_cols = [col for col in df1.columns if df1[col].dtype != 'object']
```

```python
freqgrph=df1.select_dtypes(include=['float','int64'])
```

```python
freqgrph1=df2.select_dtypes(include=['float','int64'])
```

```python
for i in num_cols:
    Q1=df1[i].quantile(0.25)
    Q2=df1[i].quantile(0.50)
    Q3=df1[i].quantile(0.75)
    iqr=Q3-Q1
    lower=Q1-(1.5*iqr)
    upper=Q3+(1.5*iqr)
    out=df1[(df1[i]<lower) | (df1[i]>upper)]
    print(f'Column: {i}, Number of outliers: {(len(out) / len(df1)) * 100}')
```

```python
for i in ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']:
    Q1=df2[i].quantile(0.25)
    Q2=df2[i].quantile(0.50)
    Q3=df2[i].quantile(0.75)
    iqr=Q3-Q1
    lower=Q1-(1.5*iqr)
    upper=Q3+(1.5*iqr)
    out1=df2[(df2[i]<lower) | (df2[i]>upper)]
    print(f'Column: {i}, Number of outliers: {(len(out1) / len(df2)) * 100}')
```

```python
#handling outliers
for column in num_cols:
        median = df1[column].median()
        df1[column] = np.where((df1[column] < lower) | (df1[column] > upper), median, df1[column])
```

```python
for column in ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History']:
        median = df2[column].median()
        df2[column] = np.where((df2[column] < lower) | (df2[column] > upper), median, df2[column])
```
